```python
#Assignmment
import pandas as pd
#To load data
titanic_data = pd.read_csv('Titanic-Dataset.csv')

#Basic info
titanic_data.info()
```

```python
#Statistical summary
titanic_data.describe()
```

```python
titanic_data["Age"] = titanic_data["Age"].fillna(titanic_data["Age"].median())
```

```python
titanic_data["Cabin"] = titanic_data["Cabin"].fillna("Unknown")
```

```python
titanic_data.describe()
#Now no missing data 
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode Sex column(LabelEncoder)
label_encoder = LabelEncoder()
titanic_data['Sex'] = label_encoder.fit_transform(titanic_data['Sex'])
print(titanic_data[['Sex']].head())
```

```python
# Encode Embarked (OneHotEncoder)
titanic_data = pd.get_dummies(titanic_data, columns=['Embarked'], drop_first=True)
```

```python
titanic_data.info()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(titanic_data['Age'], bins=30, edgecolor='black')
plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
titanic_data.to_csv("titanic_cleaned.csv", index=False)
```
